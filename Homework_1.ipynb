### Задание №1 к уроку №2

Код на Python, реализующий расчет длины вектора, заданного его координатами

```python
vector_1 = [20, 20, 20]
vector_2 = [0, 0, -20]
```

```python
# Функция сложения векторов
def add_vectors(v, w):
    return [v_i + w_i for v_i, w_i in zip(v, w)]
```

```python
vector_3 = add_vectors(vector_1, vector_2)
vector_3
```

```python
# Функция получения суммы квадратов вектора
def sum_of_squares(v):
    return sum(vi ** 2 for vi in v)

# Функция вычисления длины вектора
def magnitude(v):
    return math.sqrt(sum_of_squares(v))
```

```python
print(f'Длина вектора: {magnitude(vector_3)}')
```

### Задание №3 к уроку №2

Построение графиков:
1.	окружности;
2.	эллипса;
3.	гиперболы.

```python
# Построение окружности
x = []
y1 = []
y2 = []

# Радиус
R = 1

for i in range(-100000, 100000):
    x_ = i/100000
    x.append(x_)
    y1.append(math.sqrt(pow(R,2) - pow(x_,2)))
    y2.append(-math.sqrt(pow(R,2) - pow(x_,2)))

rcParams['figure.figsize'] = 5, 5

plt.plot(x, y1)
plt.plot(x, y2)

plt.title('Графики y(x)')
plt.xlabel("Переменная x")
plt.ylabel("Функция y(x)")

plt.show()
```

```python
# Построение эллипса
x = []
y1 = []
y2 = []

# Коэффициенты
a = 2
b = 1

for i in range(-100000, 100000):
    x_ = i/50000
    x.append(x_)
    y1.append(b*math.sqrt(1 - pow(x_,2)/pow(a, 2)))
    y2.append(-b*math.sqrt(1 - pow(x_,2)/pow(a, 2)))

rcParams['figure.figsize'] = 8, 4

plt.plot(x, y1)
plt.plot(x, y2)

plt.title('Графики y(x)')
plt.xlabel("Переменная x")
plt.ylabel("Функция y(x)")

plt.show()
```

```python
# Построение гиперболы
x = []
y1 = []
y2 = []

# Коэффициенты
a = 2
b = 4

for i in range(2000, 8000):
    x_ = i/1000
    x.append(x_)
    x.append(- x_)
    y1.append(math.sqrt(pow(b, 2) * ((pow(x_, 2) / pow(a, 2)) - 1)))
    y2.append(-math.sqrt(pow(b, 2) * ((pow(x_, 2) / pow(a, 2)) - 1)))
    y1.append(math.sqrt(pow(b, 2) * ((pow(-x_, 2) / pow(a, 2)) - 1)))
    y2.append(-math.sqrt(pow(b, 2) * ((pow(-x_, 2) / pow(a, 2)) - 1)))  
    
rcParams['figure.figsize'] = 8, 10

plt.scatter(x, y1)
plt.scatter(x, y2)

plt.title('Графики y(x)')
plt.xlabel("Переменная x")
plt.ylabel("Функция y(x)")

plt.show()
```

### Задание №1 к уроку №3

Нарисуйте график функции: y(x) = k∙cos(x – a) + b

```python
rcParams['figure.figsize'] = 6, 7

x = np.linspace(-10, 10, 100)

k1, a1, b1  = 1, 0, 1
y1 = k1 * np.cos(x - a1) + b1

k2, a2, b2  = 2, 1, 2
y2 = k2 * np.cos(x - a2) + b2

plt.plot(x, y1)
plt.plot(x, y2)

plt.grid(True)
plt.title('Графики y(x)')
plt.xlabel('Переменная x')
plt.ylabel('Функция y(x)')
plt.legend(labels=['y1', 'y2'])

plt.show()
```
